```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
data = pd.read_csv("Iris.csv")
```

```python
sns.countplot(data=data, x='Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Count of Species')
plt.show()
```

```python
for feature in data.columns[:-1]:  # Exclude the last column 'Species'
    sns.kdeplot(data=data, x=feature, fill=True)
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.title(f'Kernel Density Estimation of {feature}')
    plt.show()
```

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for feature in features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=data, x=feature, kde=True)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(f'Histogram of {feature}')
    plt.show()
```

```python
data.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=data, size=5)
```

```python
sns.FacetGrid(data, hue="Species" ) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm" , s=7) \
   .add_legend()
```
